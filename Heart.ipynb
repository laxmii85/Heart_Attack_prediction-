1. Heart Disease Class Distribution

```python
plt.figure(figsize=(5, 4))
sns.countplot(x='HeartDisease', data=df, palette='Set2')
plt.title("Heart Disease Class Distribution")
plt.xlabel("Heart Disease (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

2. Numerical Features - Histogram + KDE

```python
numerical_cols = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']

plt.figure(figsize=(12, 10))
for idx, col in enumerate(numerical_cols, 1):
    plt.subplot(3, 2, idx)
    sns.histplot(df[col], kde=True, color='skyblue')
    plt.title(f'{col} Distribution')
plt.tight_layout()
plt.show()
```

3. Boxplots (Numerical vs HeartDisease)

```python
plt.figure(figsize=(12, 10))
for idx, col in enumerate(numerical_cols, 1):
    plt.subplot(3, 2, idx)
    sns.boxplot(x='HeartDisease', y=col, data=df, palette='Set3')
    plt.title(f'{col} by Heart Disease')
plt.tight_layout()
plt.show()
```

4. Categorical Features - Countplots

```python
df.info()
```

```python
categorical_cols = ['Sex', 'ChestPainType', 'FastingBS', 'ExerciseAngina', 'ST_Slope', 'RestingECG']

plt.figure(figsize=(15, 18))
for idx, col in enumerate(categorical_cols, 1):
    plt.subplot(4, 2, idx)
    sns.countplot(x=col, hue='HeartDisease', data=df, palette='pastel')
    plt.title(f'{col} vs Heart Disease')
    plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

(Optional) Pie Chart for Sex Distribution

```python
plt.figure(figsize=(5,5))
df['Sex'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'])
plt.title("Sex Distribution")
plt.ylabel('')
plt.show()
```

```python
df['Cholesterol'].value_counts()
```

```python
ch_mean = df.loc[df['Cholesterol'] != 0,'Cholesterol'].mean()
```

```python
df['Cholesterol'] = df['Cholesterol'].replace(0,ch_mean)
df['Cholesterol'] = df['Cholesterol'].round(2)
```

```python
resting_bp_mean = df.loc[df['RestingBP'] != 0, 'RestingBP'].mean()

df['RestingBP'] = df['RestingBP'].replace(0, resting_bp_mean)

df['RestingBP'] = df['RestingBP'].round(2)
```

```python
def plotting(var,num):
    plt.subplot(2,2,num)
    sns.histplot(df[var],kde = True)

plotting('Age',1)
plotting('RestingBP',2)
plotting('Cholesterol',3)
plotting('MaxHR',4)


plt.tight_layout()
```
